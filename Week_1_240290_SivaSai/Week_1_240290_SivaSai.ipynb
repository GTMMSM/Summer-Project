##### Importing Libraries

```python
import numpy as np
import random
from collections import defaultdict
import queue
```

##### Classifying and Labelling students

```python
# An empty list to store students
students = []
total_students = 1000

# storing 1000 students.
for i in range(1,1001):
    students.append(i)

# initializing no.of classes,sections.
num_classes = 10
num_sections = random.randint(4,5)

# function to classify students into clasees and sections
def classify_students(num_classes,num_sections,total_students):
    class_size = int(total_students/num_classes)
    section_size = int(class_size/num_sections)
    classes = [] # list to store classes

    # for loop to classify students into classes
    for i in range(1,num_classes+1):
        if num_sections == 4:
            for letter in ['A','B','C','D'] :
                section = str(i)+letter
                sections = [section]*section_size
                classes.extend(sections)
        else:
            for letter in ['A','B','C','D','E'] :
                section = str(i)+letter
                sections= [section]*section_size
                classes.extend(sections)
        
    # grouping students based on classes and section
    students_by_section = defaultdict(list)
    students_by_class = defaultdict(list)

    for i , student_class in enumerate(classes) :
        cls = student_class[:-1]
        student_section = student_class
        students_by_section[student_section].append(i+1)
        students_by_class[cls].append(i+1)

    return (classes,students_by_class,students_by_section)

student_groups = classify_students(num_classes,num_sections,total_students)
classes = student_groups[0]
students_by_class = student_groups[1]
students_by_section = student_groups[2]

print(classes)
```

##### Adjacency list

```python
num_edges = random.randint(950,1050)
friend_groups = defaultdict(list)
edges_added = set()

# popular students case
popular_students = set()
for i in range(1,51):
    u = random.randint(1,1000)
    popular_students.add(u)

while len(edges_added) < num_edges :
    u = random.randint(1,1000)
    section_u = classes[u - 1]
    class_u = section_u[:-1]

    r = random.random()
    candidates = []

    if r < 0.05:
        # 5% chance for same section
        for v in students_by_section[section_u]:
            if v != u :
                candidates.append(v)
    elif r < 0.08:
        # 3% chance for same class,different section
        for v in students_by_class[class_u]:
            if v != u :
                candidates.append(v)

    elif r < 0.10:
        # 2% chance for popular students
        for v in popular_students:
            if v != u :
                candidates.append(v)

    else:
        # 90% chance for friend of friends
        fof_candidates = set()
        if u in friend_groups:
            direct_friends = set([f[0] for f in friend_groups[u]])
            for friend in direct_friends:
                for fof in friend_groups[friend]:
                    fof_id = fof[0]
                    if fof_id != u and fof_id not in direct_friends:
                        fof_candidates.add(fof_id)
        candidates = list(fof_candidates)

    if not candidates:
        continue

    v = random.choice(candidates)
    edge = (min(u,v),max(u,v))
    if edge in edges_added:
        continue
    if len(friend_groups[u]) >= 8 or len(friend_groups[v]) >= 8:
        continue


    # students in same classes and sections case
    if classes[u-1][0] == classes[v-1][0]:
        if classes[u-1][1] == classes[v-1][1]:
            weight = random.randint(1,3)
        else: 
            weight = random.randint(3,6)
    else: 
        weight = random.randint(6,10)
    
    if u in popular_students or v in popular_students:
        weight = min(weight,random.randint(1,7))


    friend_groups[u].append((v,weight))
    friend_groups[v].append((u,weight))
    edges_added.add(edge)

print(friend_groups)
```

##### Analyzing friend groups

###### Using DFS to find number of friend groups

```python
n = total_students
g = friend_groups
components = [0]*(n+1)
visited = [False]*(n+1)

def dfs(node,count):
    visited[node] = True
    components[node] = count

    for neighbour in g[node]:
        neighbour_index = neighbour[0]
        if not visited[neighbour_index]:
            dfs(neighbour_index,count)

def find_components():
    count = 0
    for i in range(1,n+1):
        if not visited[i] :
            count += 1
            dfs(i,count)
    return (count,components)

Groups = find_components()
count = Groups[0]
components = Groups[1]

# To find largest group
group_count = [0] * (count+1)
group_count[0] = 9
for i in range(1,1001):
    group_index = components[i]
    group_count[group_index] += 1

# Some important informations
print(f'Total Students : {total_students}')
print(f'Total friendship bonds : {num_edges}')
print(f'Total friend groups : {Groups[0]}')

print(f'Largest group is {group_count.index(max(group_count))} and its size is {max(group_count)}')
print(f'Smallest group is {group_count.index(min(group_count))} and its size is {min(group_count)}')
```

##### Finding Closest Friendship Paths

###### Dijkstra's Algorithm

```python
student_pairs = []
for key in friend_groups :
    student_pairs.append(key)

g = friend_groups
n = total_students
friend_path = defaultdict(list)

for i in range(0,5) :
    s = student_pairs[i]
    def shortest_path(g,n,s) :
        visited = [False] * (n+1)
        dist = [float('inf')] * (n+1)
        dist[s] = 0
        pqueue = queue.PriorityQueue(maxsize=0)
        pqueue.put((s,0))
        while not pqueue.empty() :
            index, minvalue = pqueue.get()
            visited[index] = True
            if dist[index] < minvalue : continue
            for edge in g[index] :
                if visited[edge[0]] :
                    continue
                new_dist = dist[index] + edge[1]
                if new_dist < dist[edge[0]] :
                    dist[edge[0]] = new_dist
                    pqueue.put((edge[0],new_dist))
        return dist
    friend_path[i].append(shortest_path(g,n,s))    

friend_paths = defaultdict(list)

for i in range(0,5) :
    s_id = student_pairs[i]
    dist_list = friend_path[i][0]
    
    for node,distance in enumerate(dist_list) :
        if distance != float('inf') and node != s_id:
            friend_paths[s_id].append((node,distance))

print(friend_paths)
```
